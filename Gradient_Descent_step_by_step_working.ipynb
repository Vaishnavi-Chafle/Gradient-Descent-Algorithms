```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples = 4, n_features = 1, n_informative = 1, n_targets = 1, noise = 80, random_state = 13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
print(reg.intercept_)
print(reg.coef_)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color="red")
```

```python
# now apply the Gradient Descent assuming slope is constant m =  78.35
# and assume starting value for intercept b=0

y_pred = ((78.35*x) + 0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = "red",label = "OLS")
plt.plot(x,y_pred,color = "#00a066",label = "b=0")
plt.legend()
plt.plot()
```

```python
# iteration 1
m = 78.35
b = 0

slope = -2 * np.sum(y - m*x.ravel() - b)  # .ravel()  is used to flatten a multi-dimensional array into a 1D array. It returns a contiguous flattened array without modifying the original array.
slope
```

```python
lr = 0.1

step_size = slope * lr
step_size
```

```python
# calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * x) + b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = "red",label = "OLS")
plt.plot(x,y_pred,color="#00a066",label = "b=0")
plt.plot(x,y_pred1,color = "#aa0066",label = "b={}".format(b))
plt.legend()
plt.plot()
```

```python
# Iteration2
slope = -2 *np.sum(y - m*x.ravel() - b)
slope
```

```python
step_size = slope *lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * x) + b)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color="red",label = "OLS")
plt.plot(x,y_pred,color = "#00a066",label="b=0")
plt.plot(x,y_pred1,color = "#aa0066",label = "b={}".format(b))
plt.plot(x,y_pred2,color="#234466",label = "b={}".format(b))
plt.legend()
plt.plot()
```

```python
# Iteration3
slope = -2 * np.sum(y - m*x.ravel() - b)
slope
```

```python
step_size = lr * slope
step_size
```

```python
b = b - step_size
b
```

```python
y_pred3 = ((78.35 * x) + b)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color="red",label = "OLS")
plt.plot(x,y_pred,color = "#00a066",label="b=0")
plt.plot(x,y_pred1,color = "#aa0066")
plt.plot(x,y_pred2,color="#234466")
plt.plot(x,y_pred3,color="violet",label = "b={}".format(b))
plt.legend()
plt.plot()
```

# **Using for loop**

```python
b = -1
m = 78.35
lr = 0.1

num_epochs = 10

for i in range(num_epochs):
  slope = -2 * np.sum(y - m*x.ravel() - b)
  b = b - (lr*slope)
  y_pred = x*m + b

  plt.plot(x,y_pred)

plt.scatter(x,y)
```

```python
b = -100
m = 78.35
lr = 0.01    # Due to the small learning rate the model may not coverge

num_epochs = 10

for i in range(num_epochs):
  slope = -2 * np.sum(y - m*x.ravel() - b)
  b = b - (lr*slope)
  y_pred = x*m + b

  plt.plot(x,y_pred)

plt.scatter(x,y)
```

```python
b = -100
m = 78.35
lr = 0.01

num_epochs = 100  # Due to the high number of epochs and small learning rate the model take too much time to converge

for i in range(num_epochs):
  slope = -2 * np.sum(y - m*x.ravel() - b)
  b = b - (lr*slope)
  y_pred = x*m + b

  plt.plot(x,y_pred)

plt.scatter(x,y)
```
